```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import io
from skimage.feature import canny, peak_local_max
from skimage.filters import roberts, sobel, threshold_otsu, threshold_yen, threshold_local, gaussian, try_all_threshold
from skimage.transform import hough_ellipse, hough_circle, hough_circle_peaks
from skimage.draw import ellipse_perimeter, circle_perimeter
from skimage.morphology import binary_erosion, binary_dilation, binary_closing, binary_opening, watershed, square
from skimage.measure import label, regionprops

from scipy import ndimage as ndi
```

```python
class Celula(object):
    def __init__(self, x0, y0, area, theta, region, a, b):
        self.x0 = int(round(x0))
        self.y0 = int(round(y0))
        self.area = area
        self.theta = theta
        self.region = region
        self.a = a
        self.b = b
        
    def inEllipse(self, x0, y0):
        minr, minc, maxr, maxc = self.region
        if ((x0 >= minc) & (x0 <= maxc)) & ((y0 >= minr) & (y0 >= maxr)):
            return True
        else: return False
    
    def getEllipse(self, npoints = 50):
        alpha = np.linspace(0, 2 * np.pi, npoints)
        theta = np.deg2rad(self.theta)
        x = self.a * np.cos(alpha) * np.cos(theta) - self.b * np.sin(alpha) * np.sin(theta) + self.x0
        y = self.a * np.cos(alpha) * np.sin(theta) + self.b * np.sin(alpha) * np.cos(theta) + self.y0
        
        return x, y
    
    def getBox(self):
        if self.region != None:
            minr, minc, maxr, maxc = self.region
            bx = (minc, maxc, maxc, minc, minc)
            by = (minr, minr, maxr, maxr, minr)
            return bx, by
        else:
            return None
        
    def center(self):
        return self.x0, self.y0
    
    def __repr__(self):
        return "(%d, %d), %d px^2" % (self.x0, self.y0, self.area)
```

# Get Cells

```python
def getCells(image, blurry_std = 4, erode_kernel = 8, watershed_kernel = 25):
    blurred = gaussian(image, sigma = blurry_std)
    edges = sobel(blurred)

    local_thresh = threshold_yen(edges)
    binary = (edges > local_thresh) * 1

    eroded = binary_erosion(binary, square(erode_kernel))

    markers = np.zeros_like(image)

    markers[image < 1] = 1
    markers[image > 10] = 2

    regions = watershed(eroded, markers) - 1
    
    splitted = regions - eroded
    splitted[splitted <= 0] = 0
    
    splitted = gaussian(splitted, sigma = 2)
    r = splitted.max() - splitted.min()
    splitted = splitted > (r * 0.2 + splitted.min()) 
    
    distance = ndi.distance_transform_edt(splitted)
    local_maxi = peak_local_max(distance, indices=False, footprint=np.ones((m, m)),
                                labels = splitted)
    markers = ndi.label(local_maxi)[0]
    labels = watershed(-distance, markers, mask = splitted)

    regions = regionprops(labels, coordinates = 'xy')
    regions = [region for region in regions if region.area >= 100]

    celulas = [Celula(*(region.centroid[::-1]), region.area, -np.rad2deg(region.orientation), region.bbox,
                      region.major_axis_length / 2, region.minor_axis_length / 2) for region in regions]
    
    return celulas
```

```python
fig, ax = plt.subplots(figsize = (1.5 * n, 1.5 * n))

plt.imshow(images[0], cmap = 'Greys')

celulas = getCells(images[0])

for i, celula in enumerate(celulas):
    x, y = celula.getEllipse()
    x0, y0 = celula.center()
    c = ax.plot(x, y)
    
    ax.text(x0, y0, i, color = 'red', fontsize = 6)

ax.set_axis_off()
```
