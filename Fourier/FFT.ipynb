```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib
```

```python
im = plt.imread("corte.jpg").sum(axis = 2) * 255 / im.max()
im.shape
```

```python
plt.imshow(im)
```

```python
ft = np.fft.fft2(im)
# plt.imshow(abs(ft))
lft = np.log(abs(ft) + 1)

th = 0

nft = np.zeros(lft.shape)
nft[lft > th] = 1

fft = ft.copy()
fft[lft > th] = 0

image = np.fft.ifft2(fft).real

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(16, 9))

ax1.imshow(im)
ax2.imshow(lft)
ax3.imshow(nft)
ax4.imshow(image, vmin=0, vmax=1)

for ax in [ax1, ax2, ax3, ax4]:
    ax.axis('off')
    
fig.tight_layout()
```
